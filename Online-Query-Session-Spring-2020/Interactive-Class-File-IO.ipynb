```python
words = list()
word_dict = dict()
```

```python
with open('File IO/pray-urdu.txt', 'r') as file:
    for line in file:
        words += line.strip().split()
print(words)
```

```python
temp = { l:[ w for w in words if len(w) == l ] for l in range(1,21) }
word_dict = { k:v for k,v in temp.items() if len(v) != 0 }
print(temp)
print(word_dict)
print(type(word_dict))
```

```python
with open('spc.txt', 'w', encoding='utf-8') as w:
    w.write(str(word_dict))
```

```python
with open('spc.txt', 'r', encoding='utf-8') as r:
    content = r.read()
    print(content)
    d = dict(content)
    print(type(d))
```

# JSON

write == dump

read == load

dump(...), dumps(...)

load(...), loads(...)

```python
import json
```

```python
with open('pre-processed-data.json', 'w', encoding='utf-8') as j:
    json.dump(word_dict,j, ensure_ascii=False)
```

```python
with open('pre-processed-data.json', 'r', encoding='utf-8') as j:
    c = json.load(j)
```

```python
print(type(c))
```

```python
print(c)
```

```python
print(json.dumps(word_dict, ensure_ascii=False))
```
